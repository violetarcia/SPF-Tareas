```python
import pandas as pd
import numpy as np 
import scipy.stats as st
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.stats.diagnostic as smd
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
```

```python
#Funciones

#Funcion de multiples graficas lineales
def Multi_Graph_plot(datos, etiquetas, titulo, y_label):
    [plt.plot(datos.iloc[:,i], label = etiquetas[i]) for i in range(len(datos.iloc[0]))]
    plt.title(titulo)   # Establece el título del gráfico
    plt.xlabel("Time")   # Establece el título del eje x: abscisa
    plt.xticks(rotation=90)
    plt.ylabel(y_label)   # Establece el título del eje y: ordenada
    plt.legend(loc='upper left') 
    
#Funcion de 3 subplots
def Tres_hist(data1, name1, data2, name2, data3, name3):
    plt.figure(figsize=[16,4])
    plt.subplot(131)
    plt.hist(data1,15,density=True,label=name1)
    plt.legend()
    
    plt.subplot(132)
    plt.hist(data2,15,density=True,label=name2)
    plt.legend()

    plt.subplot(133)
    plt.hist(data3,15,density=True,label=name3)
    plt.legend()
    #np.histogram
    plt.show()

#Funcion histograma
def Graph_hist(datos, titulo, y_label):
    [freq,x,p] = plt.hist(datos, density = True, alpha = 0.3, bins = 40)
    plt.title(titulo)   # Establece el título del gráfico
    plt.xlabel("Time")   # Establece el título del eje x: abscisa
    plt.ylabel(y_label)   # Establece el título del eje y: ordenada
    plt.show()
    return freq, x, p

#Funcion de regresion lineal
def reglin(x, y, name):
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    B0 = model.params[0]
    B1 = model.params[1]
    x = x[:, 1]
    
    #Grafica
    x2 = np.linspace(x.min(), x.max(), 100)
    y_hat = x2*B1 + B0
    plt.scatter(x, y, alpha = 1)
    plt.plot(x2, y_hat, 'r')
    plt.title(name)
    plt.xlabel('X Value')
    plt.xlabel('Y Value')
    return model, B0, B1

def Kolmogrov_Test(data, dist_name):
    dist = getattr(st, dist_name)
    param = dist.fit(data)
    D, p = st.kstest(data, dist_name, args=param)
    print("p value for " + dist_name + " = "+str(p))


def Quant_Quant(Y, best_dist):
    dist = getattr(st, best_dist)
    param = dist.fit(Y)
    (slope, intercept) = st.probplot(Y, dist=best_dist, sparams=param, fit=True, plot=plt)
    
    u = np.random.rand(len(Y))
    # Comparación de quantiles númericos
    Y_n = dist.ppf(u, *param[:-2], loc=param[-2], scale=param[-1])
    q1 = np.percentile(Y,[25,50,75])
    q2 = np.percentile(Y_n,[25,50,75])
    print('Empirical quantile: ',q1)
    print('Theorical quantile: ',q2)
    
def Hist_emp_Best_dist_line(Y, best_dist, title):
    dist = getattr(st, best_dist)
    x = np.linspace(min(Y),max(Y), len(Y))
    param = dist.fit(Y)
    pdf_fitted = dist.pdf(x, *param[:-2], loc=param[-2], scale=param[-1])
    
    #Graficas
    plt.plot(x,pdf_fitted)
    plt.hist(Y, alpha = 0.5, density = True, bins=15)
    plt.title(title) 
    

def Chi_Cuadr(Y, best_dist):
    dist = getattr(st, best_dist)
    J = len(Y)
    
    #Calcular la frecuecia sin graficar el histograma
    hist, bin_edges = np.histogram(Y, bins=(J))
    bins = bin_edges[1]-bin_edges[0]
    freq = (hist/(bins*sum(hist)))
    
    x = np.linspace(min(Y),max(Y), J)
    
    param = dist.fit(Y)
    pi = dist.pdf(x, *param[:-2], loc=param[-2], scale=param[-1])
    
    Y_np = np.array(Y)
    Ei = Y_np*pi
    # Calculo usando la librería estadística de la chi cuadrada

    X_chi = st.chisquare(freq,Ei)
    print('Valor de chi cuadrado librería = ',X_chi)
    
def calc_KDE(data, title):
    #Calculat KDE
    kde = sm.nonparametric.KDEUnivariate(data)
    kde.fit(kernel='gau', bw='silverman')
    #Graph
    plt.plot(kde.support, kde.density, lw=3, label='KDE from samples', zorder=10)
    plt.hist(data, density = True, alpha = 0.4, bins = 5)
    plt.title(title)
    return kde

def Reg_Log(X, Y, ngrado):
    poly=PolynomialFeatures(ngrado)
    Xa=poly.fit_transform(X)
    
    logreg=linear_model.LogisticRegression(C=1e20)#
    logreg.fit(Xa,Y)
    Yg=logreg.predict(Xa)
    
    return logreg.fit(Xa,Y), Yg
```

**A continuacion se utilizo es: scipy.stats.kstest**

**Esto realiza una prueba de la distribución G (x) de una variable aleatoria observada contra una distribución dada F (x). Bajo la hipótesis nula, las dos distribuciones son idénticas, G (x) = F (x). La hipótesis alternativa puede ser "de dos lados" (por defecto), "menos" o "mayor". La prueba KS solo es válida para distribuciones continuas.**
