<font color= green > 1. **_Find all of the numbers from 1-1000 that are divisible by 7._**


```python
list = [x for x in range(1000) if x%7 == 0]
print(list)
```

<font color= green > 2. **_Find all of the numbers from 1-1000 that have a 3 in them._**


```python
list = [x for x in range(1000) if '3' in str(x)]
print(list)
```

<font color= green > 3. **_Count the number of spaces in a string._**


```python
def spaces(list):
    ans = [x for x in list if str(x) == ' ']
    return len(ans)

lista = "Hello world ! !"
space = spaces(lista)
print(space)
```

<font color= green > 4. **_Remove all of the vowels in a string [make a list of the non-vowels]._**


```python
def noVowels(string):
    vowels = ('A', 'E', 'I', 'O', 'U')
    ans = [x for x in string if x.upper() not in vowels]
    return ans

lista = "Hello world"
cons = noVowels(lista)
print(cons)
```

<font color= green > 5. **_Find all of the words in a string that are less than 4 letters._**


```python
def shortWords(string):
    string = string.split()
    ans = [x for x in string if len(x) < 4]
    return ans

lista = "Una niña que juega"
words = shortWords(lista)
print(words)
```

<font color= green > 6. **_Use a dictionary comprehension to count the length of each word in a sentence._**


```python
def lenWord(string):
    string = string.split()
    ans = [len(x) for x in string]
    return ans

lista = "Hellow world"
lenLista = lenWord(lista)
print(lenLista)
```

<font color= green > 7. **_Use a nested list comprehension to find all of the numbers from 1-1000 that are divisible by any single digit besides 1 (2-9)._**

```python
results = [[x for x in range(1,1001) if x % divisor == 0] for divisor in range(2,10)]
print(results)
```
